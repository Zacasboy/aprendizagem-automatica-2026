```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
optical_recognition_of_handwritten_digits = fetch_ucirepo(id=80)

# data (as pandas dataframes)
X = optical_recognition_of_handwritten_digits.data.features
y = optical_recognition_of_handwritten_digits.data.targets

# metadata
print(optical_recognition_of_handwritten_digits.metadata)

# variable information
print(optical_recognition_of_handwritten_digits.variables)
##data = pd.read_csv("URL", sep=";")

print(X)
```

**Dados de Treino**

```python
#Imports
import pandas as pd
import pickle as p1
from sklearn import linear_model
import numpy as np
```

```python
#Código principal
M=len(X)
print("nº total de dados",len(X)) #nº total de dados
print("nº total de parâmetros",len(np.array(X[:1])[0])," + 1")

m=int(2*M/3)  #3500 nº de dados de treino
print("nº total de dados de treino",m)

train_data=X[:m]
data_X=X[:m]
data_Y=y[:m]

# Restantes dados de teste
datateste_X=X[m:]
datateste_Y=y[m:]
Dadosteste=[datateste_X,datateste_Y]
with open("Dadosteste.pkl", "wb") as Dt:
    p1.dump(Dadosteste, Dt)


#print(data_X)
#print(data_Y)
from sklearn.linear_model import LogisticRegression
regr = LogisticRegression(max_iter=1000, multi_class='ovr')
Modelo = regr.fit(data_X, data_Y.values.ravel())  # ravel() para formato correto


# Guarda o modelo usando 'with'
with open('NDigitsPredictor', 'wb') as PickleModelo:
    p1.dump(Modelo, PickleModelo)
print("NDigitsPredictor guardado com sucesso!")
```

**Dados De Teste - N Dígitos**

```python
with open("Dadosteste.pkl", "rb") as Dt:
  Dadosteste=p1.load(Dt)

data_teste_X=Dadosteste[0]
data_teste_Y=Dadosteste[1]

print(type(data_teste_X))
# Carregar modelo
Modelogravado = p1.load(open('NDigitsPredictor', 'rb'))
#print("Coeficientes:\n", Modelogravado.coef_)

# Fazer predições
y_pred = Modelogravado.predict(data_teste_X)

# Para classificação, precisamos arredondar as predições para o dígito mais próximo
# e limitar entre 0 e 9 (dígitos válidos)
right = np.sum(y_pred == data_teste_Y.values.ravel())
wrong = len(data_teste_Y) - right
total = len(data_teste_Y)


print(f"Eficácia de acertos:{((right/total)*100):.2f}")
print(f"Taxa de erro: {((wrong/total)*100):.2f}")
```
